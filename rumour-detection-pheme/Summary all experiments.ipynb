### Summary - Experiments with tabular dataset

#### Xgboost - SMOTE oversampling - Filtering data with time

Xgboost SMOTE Filter Node 2024-09-30 10 min-3days

* Filtering both the post publish time (minutes since first post) and reply time (minutes since post publication)
* SMOTE to oversample the minory class
* Time cute range from 10 minutes to 3 days
* Ex: 50 minutes will pick tweets published 50 minutes after first tweet considering 50 minutes of interactions for both training and test set 

```python

# Get the experiment ID or name
df = pd.read_csv(r"Xgboost SMOTE Filter Node 2024-09-30 10 min-3days")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert time_cut to hours (assuming it's in seconds)
df['metrics.time_cut_hours'] = df['metrics.time_cut'] / 60

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plotting Time Cut (in hours) vs. Test Recall in the first subplot
sns.lineplot(ax=axes[0], x='metrics.time_cut_hours', y='metrics.test_recall', data=df[df['metrics.time_cut_hours']<=1e10])
axes[0].set_title('Time Cut (Hours) vs. Test Recall')
axes[0].set_xlabel('Time Cut (Hours)')
axes[0].set_ylabel('Test Recall')
axes[0].grid(True)

# Plotting Time Cut (in hours) vs. Test Precision in the second subplot
sns.lineplot(ax=axes[1], x='metrics.time_cut_hours', y='metrics.test_precision', data=df[df['metrics.time_cut_hours']<=1e10])
axes[1].set_title('Time Cut (Hours) vs. Test Precision')
axes[1].set_xlabel('Time Cut (Hours)')
axes[1].set_ylabel('Test Precision')
axes[1].grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
